```python
def wishMe():
    hour = datetime.datetime.now().hour
    
    if hour >= 0 and hour < 12:
        speak("Good morning")
        print("Good morning")
    elif hour >= 12 and hour < 16:
        speak("Good Afternoon")
        print("Good Afternoon")
    elif hour >= 16 and hour < 19:
        speak("Good evening")
        print("Good evening")
    elif hour >= 19 and hour < 23:
        speak("Before going to night how can I help you")
        print("Before going to night how can I help you")
    else:
        speak("How can I help you")
        print("How can I help you")
```

```python
def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = r.listen(source)

        try:
            print("Recognizing...")
            statement = r.recognize_google(audio, language="en-in")
            print(f"User said: {statement}\n")
            return statement.lower()
        except Exception as e:  
            print(e)
            speak("Pardon me, please repeat")
            return "None"
```

```python
from googletrans import Translator

def translate_text_to_all_regional_languages(text):
    translator = Translator()
    regional_languages = ['hi', 'bn', 'te', 'mr', 'ta', 'gu', 'kn', 'ur', 'pa', 'as', 'or', 'ml']  # Language codes for Hindi, Bengali, Telugu, Marathi, Tamil, Gujarati, Kannada, Urdu, Punjabi, Assamese, Oriya, and Malayalam
    translations = {}

    for lang_code in regional_languages:
        translated_text = translator.translate(text, dest=lang_code)
        translations[lang_code] = translated_text.text

    return translations

# Example usage
original_text = "Hello, how are you?"  # English
translations = translate_text_to_all_regional_languages(original_text)

for lang_code, translated_text in translations.items():
    print(f"{lang_code}: {translated_text}")
```
